# Cleaning data
We will now enter the last process in data wrangling which is data cleaning. In this stage, we will clean up the various problems that have been identified in the data assessing process.

```python
customer_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi sekarang: ", customer_df.duplicated().sum())
```

```python
customer_df[customer_df.gender.isna()]
```

```python
customer_df.gender.value_counts()
```

```python
customer_df.fillna(value="Prefer not to say", inplace=True)
```

```python
customer_df.isna().sum()
```

```python
customer_df[customer_df.age == customer_df.age.max()]
```

```python
customer_df.replace(customer_df.age.max(), 70, inplace=True)
```

```python
customer_df[customer_df.age == customer_df.age.max()]
```

```python
customer_df.replace(customer_df.age.max(), 50, inplace=True)
```

```python
customer_df[customer_df.age == customer_df.age.max()]
```

```python
customer_df.describe()
```

```python
datetime_column = ["order_date", "delivery_date"]

for column in datetime_column:
    orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

```python
product_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi sekarang: ", product_df.duplicated().sum())
```

```python
sales_df[sales_df.total_price.isna()]
```

```python
sales_df["total_price"] = sales_df["price_per_unit"]* sales_df["quantity"]
```

```python
sales_df.isna().sum()
```

# EDA

```python
customer_df.describe(include="all")
```

```python
customer_df.groupby(by="gender").agg({"customer_id": "nunique", "age": ["max", "min", "mean", "std"] })
```

```python
customer_df.groupby(by="city").customer_id.nunique().sort_values(ascending=False)
```

```python
customer_df.groupby(by="state").customer_id.nunique().sort_values(ascending=False)
```

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
```

```python
orders_df.info()
```

```python
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
```

```python
orders_df["delivery_time"] = round(delivery_time/86400)
orders_df
```

```python
orders_df.describe(include="all")
```

```python
customer_id_in_orders_df = orders_df.customer_id.tolist()
```

```python
customer_df["status"] = customer_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
```

```python
customer_df.sample(5)
```

```python
customer_df.groupby(by="status").customer_id.count()
```

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customer_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

```python
orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

```python
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
```

```python
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
product_df.describe(include="all")
```

```python
sales_df.describe(include="all")
```

```python
product_df.sort_values(by="price", ascending=False)
```

```python
product_df.groupby(by="product_type").agg({
    "product_id": "nunique",
    "quantity": "sum",
    "price": ["min", "max"]
})
```

```python
product_df.groupby(by="product_name").agg({
    "product_id": "nunique",
    "quantity": "sum",
    "price": ["min", "max"]
})
```

```python
sales_product_df = pd.merge(
    left=sales_df,
    right=product_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
sales_product_df.head()
```

```python
sales_product_df.groupby(by="product_type").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

```python
all_df = pd.merge(
    left=sales_product_df,
    right=orders_customers_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
all_df.head()
```

```python
all_df.groupby(by=["state", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
all_df.groupby(by=["gender", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
all_df.groupby(by=["age_group", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```
